```python
# -*- coding: utf-8 -*-
import numpy as np

import rampy as rp

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,4000,1.0)
noise = 20.
y1 = rp.gaussian(x,400,500,100) + rp.gaussian(x,200,800,500) + rp.gaussian(x,500,1200,100) + rp.gaussian(x,1000,3300,250)+np.random.randn(len(x))*noise
y2 = rp.gaussian(x,400,500,100) + rp.gaussian(x,200,800,500) + rp.gaussian(x,500,1200,100) + rp.gaussian(x,1600,3300,250)+np.random.randn(len(x))*noise
y3 = rp.gaussian(x,400,500,100) + rp.gaussian(x,200,800,500) + rp.gaussian(x,500,1200,100) + rp.gaussian(x,2000,3300,250)+np.random.randn(len(x))*noise

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

```
